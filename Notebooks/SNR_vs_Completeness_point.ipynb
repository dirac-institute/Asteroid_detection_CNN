```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../DATA/point_sources/train.csv")
df.head(10)
```

```python
snr = df["SNR_estimation"]
stack_detected = df["stack_detection"]
det, bin_edges = np.histogram(snr[stack_detected], bins=80, range= (2, 12))
all_det, bin_edges = np.histogram(snr, bins=bin_edges, range= (2, 12))
```

```python
plt.bar(bin_edges[:-1], det/all_det)
plt.xlim(2,12)
plt.ylim(0,1)
plt.xticks((np.arange(2,12+1, 1)))
plt.yticks((np.arange(0,1+0.1, 0.1)))
plt.hlines(0.5,2,12, color="r", linestyles="--")
plt.vlines(5,0,1, color="r", linestyles="--")
plt.ylabel("Completeness")
plt.xlabel("SNR")
```
