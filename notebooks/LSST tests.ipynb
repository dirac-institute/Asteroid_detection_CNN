```python
min_ra = 1000
min_dec = 1000
max_ra = 0
max_dec= 0
registry = butler.registry
#for col in registry.queryCollections():
#    print(col)
for ref in registry.queryDatasets('calexp', collections=OUTPUT_COLL, instrument='HSC'):
    injected_calexp = butler.get(
        "injected_calexp",
        dataId=ref.dataId,
        collections=OUTPUT_COLL,
    )
    start = injected_calexp.wcs.pixelToSky(0,0)
    end = injected_calexp.wcs.pixelToSky(injected_calexp.getDimensions()[0],injected_calexp.getDimensions()[1])
    min_ra = min([min_ra, start.getRa()])
    min_dec = min([min_dec, start.getDec()])
    max_ra = max([max_ra, end.getRa()])
    max_dec = max([max_dec, end.getDec()])
    print ("ra:", start.getRa(), end.getRa(), "dec:", start.getDec(), end.getDec())
    
```

```python
import matplotlib.pyplot as plt
from lsst.afw.display import getDisplay, setDefaultBackend

# Set matplotlib as the default backend.
setDefaultBackend("matplotlib")

# Create a figure and a display.
fig, ax = plt.subplots(figsize=(6,8))
display = getDisplay(frame=fig, reopenPlot=True)

# Display the injected_postISRCCD image.
display.mtv(injected_calexp [:, 3500:])
```

```python
# Sort the mask plane dict by bit value and print their mask plane color.
mask_plane_dict = injected_calexp.mask.getMaskPlaneDict()
sorted_mask_plane_dict = dict(
    sorted(
        mask_plane_dict.items(),
        key=lambda item: item[1]
    )
)
max_key_length = max(map(len, sorted_mask_plane_dict.keys()))
for key in sorted_mask_plane_dict.keys():
    print(f"{key: <{max_key_length}} : {display.getMaskPlaneColor(key)}")
```
